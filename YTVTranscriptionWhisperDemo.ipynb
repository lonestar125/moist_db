```python
import csv
import whisper
from pytube import YouTube
import os

model = whisper.load_model("base")

with open("checkpoint.csv", "r", encoding="utf8") as f:
    resultreader = csv.reader(f)
    result = [el for el in resultreader]

with open("moist_shortened.csv", "r", encoding="utf8") as f:
    csvreader = csv.reader(f)
    data = [el for el in csvreader]

links = [el[0] for el in data]

fail_count = 0
check = 0
current = len(result)
for el in links[current:]:
    try:
        yt = YouTube(el)
        video = yt.streams.filter(only_audio=True).first()
        video = video.download("", "audio.mp3")
        transcript = model.transcribe("audio.mp3", fp16=True, language="English")
        result.append([transcript['text']])
    except:
        try:
            yt = YouTube(el)
            video = yt.streams.filter(only_audio=True).first()
            video = video.download("", "audio.mp3")
            transcript = model.transcribe("audio.mp3", fp16=True, language="English")
            result.append([transcript['text']])
        except:
            fail_count += 1
            print(el)
    if links.index(el) % 1 == 0:
        check += 1
        with open("checkpoint.csv", "w", encoding="utf8", newline="") as f:
            writer = csv.writer(f, delimiter="|")
            writer.writerows(result)
        print("saved: ", check)


```
